# One layer analysis - Cu sample

## load the s parameter

```python
def ufloat_converter(string):
    parts = string.split('+/-')
    if len(parts) == 2:
        return ufloat(float(parts[0]), float(parts[1]))
    else:
        return None
    
s_parm_measurement = pd.read_csv('./libs/s_parameter_measurement/cu_500_s', names=['energy', 's'], skiprows=1, index_col=0, converters={'s': ufloat_converter})
s_parm_measurement = s_parm_measurement['s']
s_parm_measurement[0.5]

energies = s_parm_measurement.index
```

```python
s_parm_measurement_nominal = [nominal_value(s) for s in s_parm_measurement]
s_parm_measurement_dev = [std_dev(s) for s in s_parm_measurement]

plt.errorbar(x=s_parm_measurement.index, y=s_parm_measurement_nominal, yerr=s_parm_measurement_dev, ls='')
plt.title('measurment S parameter vs the positrons energy, Cu 250 deg')
plt.grid(True)
plt.xlabel('Energy[keV]')
plt.ylabel('S Parameter')
plt.legend(['measurment', 'fit'])
```

## data for the bulk diffusion length optimization in annheled copper

physical parameters and initial sample 

```python
# This is the known diffusion length in the literature
L1 = 126*(nano/micro) # micrometers
D = 1 * (centi/micro) ** 2  # 1 cm**2 /second
lambda_b1 = D/L1**2
alpha = D/(10 * (nano/micro)) # assuming around 10 nm

material_guess = Material(D,0,annihilation_rate_bulk=lambda_b1)
one_layer = Layer(0, 15, material_guess)
initial_sample = Sample([one_layer], alpha)

copper_density = 8.96 # gr/cm**
```

implementation profile calculation

```python
implementation_depth = np.linspace(0, 15, 10000)
implementation_profiles = [ghosh_profile(implementation_depth, energy, copper_density, data_gos.iloc[9]) for energy in energies]
```

fitting

```python
vedbfit = OneBulkDiffusionLengthOptimization(positron_implementation_profiles=implementation_profiles,
                                             s_measurement=s_parm_measurement,
                                             initial_guess=initial_sample,
                                             num_of_mesh_cells=10000)
parm = vedbfit.optimize_diffusion_length(bounds=[(0, 0), (np.inf, np.inf)])
parm
```

### nice fit, poor measurement
indeed the value for the copper diffusion length is around 130+-20

```python
# find s_parm using linear regression
effective_rate_fit = 1/nominal_value(parm['layer_0']) ** 2
effective_surface_rate = 1/nominal_value(parm['surface'])
sample = vedbfit.make_sample(effective_surface_rate , effective_rate_fit)

annihilation_channel_rate_matrix = vedbfit.rate_matrix(sample)
s_vec = np.linalg.lstsq(annihilation_channel_rate_matrix, vedbfit.s_measurement, rcond=None)[0]
s_sample = annihilation_channel_rate_matrix @ s_vec
        
plt.errorbar(x=energies, y=vedbfit.s_measurement, yerr=vedbfit.s_measurement_dev, ls='')
plt.plot(energies, s_sample, '.')
plt.title('artificial S parameter vs the positrons energy')
plt.grid(True)
plt.xlabel('Energy[keV]')
plt.ylabel('S Parameter')
plt.legend(['data', 'fit'])
```
